```python
import pandas as pd
import quandl
import datetime
import numpy as np
import warnings      
warnings.filterwarnings('ignore')
```

```python
# import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
start_date, end_date = datetime.datetime(2010,1,1), datetime.date.today() 

apple = quandl.get("WIKI/" +  "AAPL", 
                   start_date=start_date, 
                   end_date=end_date)

apple.columns = [el.lower().replace(' ','_').replace('.','').replace('-','_') for el in list(apple.columns)]

apple.head()
```

```python
# apple["Adj. Close"].plot(grid = True);
```

```python
window = 20
apple['20_days'] = np.round(apple['adj_close'].rolling(window = window, center = False).mean(), 2)
apple.dropna(inplace=True)
```

```python
window = 50
apple['50_days'] = np.round(apple['adj_close'].rolling(window = window, center = False).mean(), 2)
apple.dropna(inplace=True)
```

```python
window = 200
apple['200_days'] = np.round(apple['adj_close'].rolling(window = window, center = False).mean(), 2)
apple.dropna(inplace=True)
```

```python
apple.head()
```

```python
apple['20_days-50_days'] = apple['20_days'] - apple['50_days']
```

```python
apple.head()
```

Now, depending if the FMA is above or below the SMA, we add a new regime column `reg`

```python
apple["reg"] = np.where(apple['20_days-50_days'] > 0, 1, 0)
apple["reg"] = np.where(apple['20_days-50_days'] < 0, -1, apple["reg"])
```

```python
apple[apple['reg']==1].head()
```

```python
apple[apple['reg']==-1].head()
```

- When FMA > SMA, a buy signal is triggered.
- When FMA < SMA, a sell signal is triggered.

Counting the buys and sells:

```python
apple["reg"].value_counts()
```

We can add a signal column:

```python
reg_orig = apple.loc[:, "reg"].iloc[-1]
apple.loc[:, "reg"].iloc[-1] = 0
apple["sign"] = np.sign(apple["reg"] - apple["reg"].shift(1))
apple.loc[:, "reg"].iloc[-1] = reg_orig
apple.dropna(inplace=True)
```

```python
apple.head()
apple.tail()
apple["sign"].value_counts()
```

Identifying the prices at every buy and every sell and creating a new `DataFrame` with trades:

```python
apple.loc[apple["sign"] == 1, "close"].head()
```

```python
apple.loc[apple["sign"] == -1, "close"].head()
```

```python
df_buy = pd.DataFrame({"price": apple.loc[apple["sign"] == 1, 'adj_close'],
                     "reg": apple.loc[apple["sign"] == 1, "reg"],
                     "sign": "buy"})

df_sell = pd.DataFrame({"price": apple.loc[apple["sign"] == -1, 'adj_close'],
                     "reg": apple.loc[apple["sign"] == -1, "reg"],
                     "sign": "sell"})

apple_trades = pd.concat([df_buy,df_sell])
apple_trades.sort_index(inplace = True)
apple_trades.head(20)
```

The profits are as follows:

```python
long_profits_price = pd.DataFrame({
        "price": apple_trades.loc[(apple_trades["sign"] == "buy") &
                                  apple_trades["reg"] == 1, "price"]})
long_profits_profit = pd.DataFrame({
        "profit": pd.Series(apple_trades["price"] - apple_trades["price"].shift(1)).loc[apple_trades.loc[(apple_trades["sign"].shift(1) == "buy") & 
                              (apple_trades["reg"].shift(1) == 1)].index].tolist()})
long_profits_end_date = pd.DataFrame({"end_date": apple_trades["price"].loc[apple_trades.loc[(apple_trades["sign"].shift(1) == "buy") & 
                                                                                    (apple_trades["reg"].shift(1) == 1)].index].index})
```

```python
long_profits = long_profits_end_date.copy()
long_profits['price'] = list(long_profits_price['price'])
long_profits['profit'] = list(long_profits_profit['profit'])
```

```python
long_profits.head()
```

### Simulated portfolio

- Total value $\$$1,000,000
- Invest only 10$\%$ of it in any trade
- Exit position if losses exceed 20$\%$ of the trade value
- Trades done in batches of 100.

```python
apple.head()
```

```python
periods = pd.DataFrame({"start": long_profits.index,
                            "end": long_profits["end_date"]})
periods.head()
```

```python
# long_profits["low"] = periods.apply(lambda x: min(apple.loc[x["start"]:x["end"], "adj_low"]), axis = 1)

long_profits["low"] = long_profits['price']
```

```python
cash = 1000000
backtest = pd.DataFrame({"Start Port. Value": [],
                         "End Port. Value": [],
                         "End Date": [],
                         "Shares": [],
                         "Share Price": [],
                         "Trade Value": [],
                         "Profit per Share": [],
                         "Total Profit": [],
                         "Stop-Loss Triggered": []})


port_value = .1  
batch = 100     
stoploss = .2    
for index, row in long_profits.iterrows():
    batches = np.floor(cash * port_value) // np.ceil(batch * row["price"]) 
    trade_val = batches * batch * row["price"] 
    if row["low"] < (1 - stoploss) * row["price"]: 
        share_profit = np.round((1 - stoploss) * row["price"], 2)
        stop_trig = True
    else:
        share_profit = row["profit"]
        stop_trig = False
    profit = share_profit * batches * batch # Compute profits
#     # Add a row to the backtest data frame containing the results of the trade
    backtest = backtest.append(pd.DataFrame({
                "Start Port. Value": cash,
                "End Port. Value": cash + profit,
                "End Date": [],
                "Shares": batch * batches,
                "Share Price": row["price"],
                "Trade Value": trade_val,
                "Profit per Share": share_profit,
                "Total Profit": profit,
                "Stop-Loss Triggered": stop_trig
            }, [index]))
    cash = np.max(0, cash + profit)
```
